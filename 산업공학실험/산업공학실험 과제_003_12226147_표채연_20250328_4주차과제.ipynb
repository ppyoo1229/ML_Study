### # 칼럼 줄이기
- 칼럼명에 중복 카테고리로 보이는 칼럼명들이 보임
- Pandas의 계산 메서드만 실습하기 때문에 칼럼을 단순화하는게 좋을 것 같음
- '해킹(계정도용)', '해킹(단순침입)', '해킹(자료유출)', '해킹(자료훼손)' -> 판다스의 sum()메서드를 사용해 '해킹'칼럼으로 만들고자 함

```python
# 해킹 관련 열 목록
해킹 = ['해킹(계정도용)', '해킹(단순침입)', '해킹(자료유출)', '해킹(자료훼손)']

#'해킹' 컬럼
df['해킹'] = df[해킹].sum(axis=1) # (axis=1) 각 행마다 합 계산 =0이나 없으면 열의 총합으로 됨

# 기존 해킹 세부 열 삭제
df.drop(columns=해킹, inplace=True)
```

```python
# 중복 카테고리 칼럼 합치기
악성프로그램 = ['악성프로그램(랜섬웨어)', '악성프로그램(기타)']
사이버사기 = ['사이버사기(직거래)', '사이버사기(쇼핑몰)', '사이버사기(게임)', '사이버사기(이메일 무역사기)', '사이버사기(기타)']
사이버금융범죄 = [
    '사이버금융범죄(피싱)', '사이버금융범죄(파밍)', '사이버금융범죄(스미싱)',
    '사이버금융범죄(메모리해킹)', '사이버금융범죄(몸캠피싱)',
    '사이버금융범죄(메신저이용사기)', '사이버금융범죄(기타)']
사이버음란물 = ['사이버 음란물(일반음란물)', '사이버 음란물(아동음란물)', '사이버 음란물(불법촬영물유포)']
사이버도박 = ['사이버 도박(스포츠토토)', '사이버 도박(경마 경륜 경정)', '사이버 도박(카지노)', '사이버 도박(기타)']

df['악성프로그램'] = df[악성프로그램].sum(axis=1)
df['사이버사기'] = df[사이버사기].sum(axis=1)
df['사이버금융범죄'] = df[사이버금융범죄].sum(axis=1)
df['사이버음란물'] = df[사이버음란물].sum(axis=1)
df['사이버도박'] = df[사이버도박].sum(axis=1)

df.drop(columns=악성프로그램, inplace=True)
df.drop(columns=사이버사기, inplace=True)
df.drop(columns=사이버금융범죄, inplace=True)
df.drop(columns=사이버음란물, inplace=True)
df.drop(columns=사이버도박, inplace=True)

df.shape
```

```python
# 순서 재정비
order = [
    '연도', '구분', '해킹', '악성프로그램', '사이버사기', '사이버금융범죄', '사이버음란물', '사이버도박',
    '서비스거부공격', '기타 정보통신망 침해형 범죄', '개인위치정보 침해', '사이버저작권침해', 
    '기타 정보통신망 이용형 범죄', '사이버 명예훼손(모욕)', '사이버 스토킹', '기타 불법 컨텐츠'
]

df = df[order]
df.head(2)
```

### # idxmax( )와 loc[ ]를 함께 이용하여 '해킹' '발생건수'가 가장 많은 '연도'찾기 

```python
# '발생건수'만 필터링
발생건수 = df[df['구분'] == '발생건수']

# 발생건수 DataFrame 안에서 idxmax 수행
idx_max = 발생건수['해킹'].idxmax()

# loc로 해킹 발생건수 '최대'의 '연도' 라벨 가져오기
발생건수.loc[idx_max, '연도']
```

### # idxmin( )와 loc[ ]를 함께 이용하여 '해킹' '발생건수'가 가장 적은 '연도'찾기

```python
idx_min = 발생건수['해킹'].idxmin()

발생건수.loc[idx_min, '연도']
```

### # 발생건수와 검거건수 총합 칼럼 추가
 - 추후 비율 계산 시 용이할 것 같음

```python
# 범죄 항목 리스트
범죄항목 = [
    '해킹', '악성프로그램', '사이버사기', '사이버금융범죄', '사이버음란물', '사이버도박',
    '서비스거부공격', '기타 정보통신망 침해형 범죄', '개인위치정보 침해', '사이버저작권침해', 
    '기타 정보통신망 이용형 범죄', '사이버 명예훼손(모욕)', '사이버 스토킹', '기타 불법 컨텐츠'
]

# '총합' 컬럼 생성
df['총합'] = df[범죄항목].sum(axis=1)

df
```

### # 검거율 계산
- 총검거건수 / 총발생건수 활용

```python
발생 = df[df['구분'] == '발생건수']
검거 = df[df['구분'] == '검거건수']

# 검거율 계산
검거율 = (검거['총합'] / 발생['총합'] * 100).round(2)
검거율
```

### # 해킹 발생 비율 계산

```python
# 해킹 비율  계산
해킹발생비율 = (발생['해킹'] / 발생['총합'] * 100).round(2)
해킹발생비율
```
